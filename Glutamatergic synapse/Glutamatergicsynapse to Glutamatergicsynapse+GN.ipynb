```python
# This program is used to relate the file that Mireia send me trough e-mail with the uniprot
# Acession code and the protein name with the gene name in Uniprot.
# A function from google is used in order to do that
########!!!!!!IT TAKES SOME TIME TO DO THE QUERY OF THE WHOLE LIST OF GENES!!!!########
# At the end we'll obtain a file with:
# UniProtKB_AC	Protein	Gene_name
# of all the genes and will be exported as (Glutamatergicsynapse+GN.txt)
# This program is also used to do the same removing duplicates (Glutamatergicsynapse+GN+Noduplicates.txt)
```

```python
# Import Pandas
import pandas as pd
```

```python
# Check how the file is processed
f = open('Glutamatergicsynapse.txt',"r")
lines = f.readlines()
f.close()
print(lines)
```

```python
# Open data into a DataFrame using Pandas
# The original file has no header
df = pd.read_csv("Glutamatergicsynapse.txt", sep="\t", header=None)
df.columns = ['UniProtKB_AC', 'Protein']
#row_numbers = df[df['UniProtKB_AC'] == 'ComplexPortal:CPX-1134'].index
df
```

```python
 AC_CODE = []
for UniProtKB_ACC in df['UniProtKB_AC']:
    mod_data = UniProtKB_ACC.replace('UniProtKB:','')
    AC_CODE.append(mod_data)
print(AC_CODE)
```

```python
# Function imported from: https://chem-workflows.com/articles/2019/10/29/retrieve-uniprot-data-using-python/
# pip install beautifulsoup4
import urllib
from bs4 import BeautifulSoup

def get_uniprot (query='',query_type='ACC'):
    #query_type must be: "PDB_ID" or "ACC"
    url = 'https://www.uniprot.org/uploadlists/' #This is the webser to retrieve the Uniprot data
    params = {
    'from':query_type,
    'to':'ACC',
    'format':'txt',
    'query':query
    }

    data = urllib.parse.urlencode(params)
    data = data.encode('ascii')
    request = urllib.request.Request(url, data)
    with urllib.request.urlopen(request) as response:
        res = response.read()
        page=BeautifulSoup(res).get_text()
        page=page.splitlines()
    return page
```

```python
gene_name = []
for AC in AC_CODE:
    if AC != 'ComplexPortal:CPX-1134':
        y=get_uniprot(query=AC ,query_type='ACC')
        for line in y:
            if 'GN   Name=' in line:
                gene_name.append(line)
    else:
        gene_name.append('NOT an AC code')
```

```python
print(gene_name)
```

```python
df_gene_name =  pd.DataFrame (gene_name, columns = ['Gene_name'])
df_gene_name[380:397]
```

```python
df['Gene_name'] = df_gene_name['Gene_name'].str.split(';').str[0]
df['Gene_name'] = df['Gene_name'].str.split('=').str[1]
df['Gene_name'] = df['Gene_name'].str.split(' ').str[0]
df[380:400]
```
