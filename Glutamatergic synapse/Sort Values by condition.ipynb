```python
# Import necessary libraries
import pandas as pd
```

```python
# Open the counts table
counts = pd.read_csv("File1_v2(Counts).txt", sep=" ", header=0)
counts[283:290]
```

```python
# sort values
counts = counts.sort_values(by = ['totals_ClinVar', 'totals_GnomAD'], axis=0, ascending=[False, True], inplace=False,
               kind='quicksort', na_position='first', ignore_index=True, key=None)
counts
```

```python
# This final table is exported as .csv
counts.to_csv('File1_v2(Counts sorted by descending Clinvar and ascending GnomAD).txt', header=True, index=None, sep=' ', mode='a')
```
