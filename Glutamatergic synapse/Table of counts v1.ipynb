```python
import pandas as pd
```

```python
ClinVar = pd.read_csv("File2_tojoinwith_GnomAD(ClinVar).txt", sep=" ", header=0)
ClinVar
```

```python
countsClinVar =ClinVar.value_counts(["Gene(s)", "Molecular Consequence", "Database"]).reset_index(name="Counts")
countsClinVar.rename(columns={'Molecular Consequence': 'Molecular Consequence/VEP Annotation'}, inplace=True)
countsClinVar
```

```python
GnomAD = pd.read_csv("File2_tojoinwith_ClinVar(GnomAD).txt", sep=" ", header=0)
GnomAD
```

```python
countsGnomAD = GnomAD.value_counts(["Gene(s)", "VEP Annotation", "Database"]).reset_index(name="Counts")
countsGnomAD.rename(columns={'VEP Annotation': 'Molecular Consequence/VEP Annotation'}, inplace=True)
countsGnomAD
```

```python
result = countsClinVar.append(countsGnomAD, ignore_index=True)
result
```

```python
result = result.sort_values(by=['Gene(s)'])
result
```

```python
result.index = range(len(result))
result
```

```python
result.to_csv('File1(Counts).txt', header=True, index=None, sep=' ', mode='a')
```
