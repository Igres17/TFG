# Import libraries

```python
import pandas as pd
import numpy as np
```

```python
# General function to open files
def open_df(file):
    df = pd.read_csv(file, delimiter = "\t")
    return(df)
```

# Pathogenic

# Frameshift

```python
df_fs_p = open_df('clinvar_result_frameshift.txt')
df_fs_p
```

```python
# Select the Gene(s) and Protein change columns to convert it into an easly understandable dataframe
df2_fs_p = df_fs_p[["Gene(s)", "Protein change"]]
df2_fs_p
```

```python
# Get the 1st Prot change and separate it into 3 new columns
df2_fs_p['Prot change'] = df2_fs_p['Protein change'].str.split(',').str[0]
df2_fs_p = df2_fs_p.drop(['Protein change'], axis=1)
df2_fs_p[['firstcol', 'secondcol', 'thirdcol']] = df2_fs_p['Prot change'].str.extract('([A-Za-z]+)(\d+\.?\d*)([A-Za-z]*)', expand = True)
df2_fs_p
```

```python
# Add the type of Protein change
df2_fs_p["Molecular Consequence"] = 'Frameshift'
df2_fs_p
```
