```python
# Import necessary libraries
import re # this library is used to delate a part of a string you don't want
# For example: gene_name=re.sub(r'{.+?}', '', gene_name) --> it deleates anything between { }

import json # this labary is usefil while saving a dictionary into a .txt file
```

```python
# This function opens a file
def open_file(file):
    file = open(file,'r')
    lines = file.readlines()
    file.close()
    return(lines)
```

```python
# This function gets the uniprot file and spit it by the genes (genes are separeted by '//\n')
def split_genes(file):
    l_gene = []
    l_final = []
    for chain in file:
        if chain != '//\n':
            l_gene.append(chain)
        else:
            l_final.append(l_gene)
            l_gene = []
    return(l_final)
```

```python
# This function creates a dictionry which relationates the gene name with all the information dic[gene_name] = info
def dic(list_genes):
    d_genes = {}
    for gene in list_genes:
        for info in gene:
            if info[0:10] == 'GN   Name=':
#                 print(info[0:20])
                semicolon = info.find(';')
                gene_name = info[10:semicolon]
                gene_name=re.sub(r'{.+?}', '', gene_name)
                gene_name=re.sub(r'{.+?,', '', gene_name)
                gene_name=re.sub(r' .+?,', '', gene_name)
                d_genes[gene_name] = gene
    return(d_genes)
```

```python
# Gets the dic and gives an output in .txt
def file_output(out_equals):
    with open('Try', 'w') as convert_file:
        convert_file.write(json.dumps(out_equals))
    return(convert_file)
```

```python
# Get a file from l_genes in .txt
def l_genes_to_output(llistafinal):
    outF = open("Gene_names.txt", "w")
    for line in llistafinal:
      # write line to output file
      outF.write(line)
      outF.write("\n") #--> si volgues separar lina per linia afegiria aix√≤
    outF.close()
```

```python
# Main Program
uniprot = open_file('uniprot-organism__Homo+sapiens+(Human)+[9606]_+reviewed_yes+anno--.txt')
genes = split_genes(uniprot)
d_genes = dic(genes)
out_dic = file_output(d_genes)


l_genes = []
for gene in d_genes:
    l_genes.append(gene)
    
out_l_gens = l_genes_to_output(l_genes)

#print(genes[0:2])
```
