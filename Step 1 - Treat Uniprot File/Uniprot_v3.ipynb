```python
# This program retrieves ["ID", "Uniprot_code", "Gene_name", "Pfam_code", "Transmembrane_Regions"] of the
# original Uniprot file and gives a list of list which contain all this information

# This list of list with Panda's help will be converted into a DataFrame and then exported as .txt file
```

```python
# Import necessary libraries
import pandas as pd
```

```python
# Creates 2 lists 1 list of lists in wich every wanted item will be added and another
# which will be used as a template to append found items in its correct position
l_all_data = [["ID", "Uniprot_code", "Gene_name", "Pfam_code", "Transmembrane_Regions", "ENST", "ENSP", "ENSG", "Canonical", "NP_XP", "NM_XM"]] 
l_ind_data = ["", "", "", "", "", "", "", "", "", "", ""] 

# Open and read the original uniprot file retrieved from uniprot
f_in = open("uniprot-organism__Homo+sapiens+(Human)+[9606]_+reviewed_yes+anno--.txt", "r")
f_data = f_in.readlines ()

# This for retrieves the information wanted accessing by startswith and using filter to clear white spaces

aux = ''
for data in f_data:
    if data.startswith("ID"):
        #print(data)
        mod_data = data.split(" ")
        mod_data = list(filter (None, mod_data)) # filter deletes all white spaces in a list
        l_ind_data[0] = mod_data[1]
    
    elif data.startswith("AC"):
        mod_data = data.split (" ")
        mod_data = list (filter(None, mod_data))
        mod_data = mod_data[1].replace(';', '')
        mod_data = mod_data.replace('\n', '')
        #print(mod_data)
        l_ind_data[1] = mod_data
    
    elif data.startswith("GN   N"):
        mod_data = data.split (" ") 
        mod_data = list (filter(None, mod_data))
        mod_data = mod_data[1].replace('Name=','')
        mod_data = mod_data.replace(';', '')
        mod_data = mod_data.replace('\n', '')
        l_ind_data[2] = mod_data
        
    elif data.startswith("DR   Pfam"):
        mod_data = data.split (" ")
        mod_data = list (filter(None, mod_data))
        mod_data = mod_data[2].replace(';', '')
        #print(mod_data)
        l_ind_data[3] = mod_data
    
    elif data.startswith("FT   TRANSMEM"):
        mod_data = data.split (" ")
        mod_data = list (filter(None, mod_data))
        mod_data = mod_data[2].replace('\n', ';')
        mod_data = mod_data.replace('..', '_')
        for e in mod_data:
            #print(e)
            aux = aux + e
        #print(laux)
        l_ind_data[4] = aux
    
    elif data.startswith("DR   Ensembl"):
        mod_data = data.split (" ")
        mod_data = list (filter(None, mod_data))
        if len(mod_data)==6 and mod_data[5].endswith("-1]\n"):
            cod_ENST = mod_data[2].replace(';', '')
            cod_ENSP = mod_data[3].replace(';', '')
            cod_ENSG = mod_data[4].replace('.', '')
            cod_ENSG = cod_ENSG.replace('\n', '')
            canonical = mod_data[5].replace('[', '')
            canonical = canonical.replace(']\n', '')
            l_ind_data[5] = cod_ENST
            l_ind_data[6] = cod_ENSP
            l_ind_data[7] = cod_ENSG
            l_ind_data[8] = canonical
        elif len(mod_data)==5:
            cod_ENST = mod_data[2].replace(';', '')
            cod_ENSP = mod_data[3].replace(';', '')
            cod_ENSG = mod_data[4].replace('.', '')
            cod_ENSG = cod_ENSG.replace('\n', '')
            l_ind_data[5] = cod_ENST
            l_ind_data[6] = cod_ENSP
            l_ind_data[7] = cod_ENSG
            l_ind_data[8] = l_ind_data[1]
    
    elif data.startswith("DR   RefSeq"):
        mod_data = data.split (" ")
        mod_data = list (filter(None, mod_data))
        #print(mod_data)
        if len(mod_data)==5 and mod_data[4].endswith("-1]\n"):
            cod_NP_XP = mod_data[2].replace(';', '')
            cod_NM_XM = mod_data[3]
            cod_NM_XM = cod_NM_XM[:-1] # is used to remove the last . of the string. In order to not use .replace who will replace the first . instead of the last one.
            #print(cod_NM_XM)
            l_ind_data[9] = cod_NP_XP
            l_ind_data[10] = cod_NM_XM
        elif len(mod_data)==4:
            cod_NP_XP = mod_data[2].replace(';', '')
            cod_NM_XM = mod_data[3].replace('.\n', '')
            #print(cod_NM_XM)
            l_ind_data[9] = cod_NP_XP
            l_ind_data[10] = cod_NM_XM
            
            

    elif data.startswith("//"): #Netejem info de 1 ind data
        #print (l_ind_data) 
        l_all_data.append(l_ind_data)
        l_ind_data = ["", "", "", "", "", "", "", "", "", "", ""] 

aux = ''

# Then pandas is used to get the information in a .txt file with 'myOutFile(Uniprot).txt' name
df=pd.DataFrame.from_records(l_all_data)
print(df)
df.to_csv('myOutFile(Uniprot).txt', header=False, index=None, sep=' ', mode='a')
```
