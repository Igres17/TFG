```python
import pandas as pd
```

```python
df_clinvar_path_names=pd.read_csv('Prot_path_clinvar.txt', delimiter='\t')
df_clinvar_path_names
```

```python
uniprot_tm = pd.read_csv('Human TM prot.tab', delimiter='\t')
uniprot_tm = uniprot_tm[['Gene names', 'Cross-reference (Pfam)']]
uniprot_tm.rename(columns = {'Gene names':'Gene(s)'}, inplace = True)
uniprot_tm
```

```python
list_genes = uniprot_tm['Gene(s)'].str.split(' ').str[0].tolist()
list_genes
```

```python
uniprot_tm['GN'] = list_genes
uniprot_tm = uniprot_tm[['GN', 'Cross-reference (Pfam)']]
uniprot_tm.rename(columns = {'GN':'Gene(s)', 'Cross-reference (Pfam)':'PFAM'}, inplace = True)
uniprot_tm
```

```python
result = pd.merge(df_clinvar_path_names, uniprot_tm, on="Gene(s)")
result
```

```python
result.to_csv('Human TM prot with path mutations', header=True, index=False, sep='\t', mode='w')
```
