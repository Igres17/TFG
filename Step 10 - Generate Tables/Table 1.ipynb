```python
import pandas as pd
```

```python
h_prot=pd.read_csv('Human prot.tab', delimiter='\t')
h_prot.dropna()
h_prot
```

```python
l_h_prot = h_prot['Cross-reference (Pfam)'].tolist()
#l_h_prot
l_pfam = []
for pfams in l_h_prot:
    split = str(pfams).split(';')
    for e in split:
        if e.startswith('PF'):
            if not e in l_pfam:
                l_pfam.append(e)
#print(l_pfam)                
print(len(l_pfam))           
```
