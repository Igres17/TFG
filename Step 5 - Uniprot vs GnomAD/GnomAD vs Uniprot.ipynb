```python
f1=open("Fitxer_intermig_2","r")
u=f1.readlines()
f1.close()
#print(u)
p=[]
header=('Gene(s) HGVSConsequence\n')
p.append(header)
for e in u:
    try:
        x=e.split()
        #print(x)
        if x[2]=='True':
            p.append(e.replace('True',''))
    except IndexError:
         pass
print(p)
```

```python
outF = open("Match(GnomAD)_nonPatho.txt", "w")
for line in p:
  # write line to output file
  outF.write(line)
  # outF.write("\n") --> si volgues separar lina per linia afegiria això
outF.close()
```
