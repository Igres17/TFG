```python
import pandas as pd
```

```python
df = pd.read_csv('all.csv', sep = ",", low_memory=False)
df = df[['Transcript','HGVS Consequence', 'ClinVar Clinical Significance']]
df = df[0:3773285]
df = df[~df['ClinVar Clinical Significance'].str.contains('Uncertain significance', na=False)]
df = df[~df['ClinVar Clinical Significance'].str.contains('Pathogenic', na=False)]
df = df[~df['ClinVar Clinical Significance'].str.contains('Conflicting interpretations of pathogenicity', na=False)]
df = df[~df['ClinVar Clinical Significance'].str.contains('Pathogenic/Likely pathogenic', na=False)]
df = df[~df['ClinVar Clinical Significance'].str.contains('Likely pathogenic', na=False)]
df = df[['Transcript','HGVS Consequence']]
df
```

```python
df2 = df[0:1853419]
df2 = df2.drop_duplicates()
i=df2[df2['Transcript']=='Transcript']
#print(i)
df2 = df2.reset_index()
df2.drop('index', axis=1, inplace=True)
df2 = df2.drop(df2.index[266])
df2 = df2.reset_index()
df2.drop('index', axis=1, inplace=True)
df2[260:270]
```

```python
df3 = df[1853419::]
df3 = df3.drop_duplicates()
df3 = df3.reset_index()
df3.drop('index', axis=1, inplace=True)
i=df3[df3['Transcript']=='Transcript']
#print(i)
df3 = df3.drop(df3.index[556])
df3 = df3.reset_index()
df3.drop('index', axis=1, inplace=True)
df3[550:560]
```

```python
################################################################################################################################
#######################################     DataFrame2    ######################################################################
################################################################################################################################
# Create a string with all the unique ENST codes without the extension (.number) in order to convert
# them into the Gene Name using a function.
u_transcripts = df2['Transcript'].unique().tolist()
#u_transcripts
sep = '.'
stripped =''
for e in u_transcripts:
    stripped = stripped + ',' +(e.split(sep, 1)[0])
stripped_df2 = stripped [1::]
print(stripped_df2)
```

```python
################################################################################################################################
#######################################     DataFrame3    ######################################################################
################################################################################################################################
# Create a string with all the unique ENST codes without the extension (.number) in order to convert
# them into the Gene Name using a function.
u_transcripts = df3['Transcript'].unique().tolist()
#u_transcripts
sep = '.'
stripped =''
for e in u_transcripts:
    stripped = stripped + ',' +(e.split(sep, 1)[0])
stripped_df3 = stripped [1::]
print(stripped_df3)
```

```python
################################################################################################################################
#######################################     DataFrame2    ######################################################################
################################################################################################################################

geneList_df2 = list(stripped_df2.split(',')) 
print(geneList_df2)
```

```python
################################################################################################################################
#######################################     DataFrame3    ######################################################################
################################################################################################################################

geneList_df3 = list(stripped_df3.split(',')) 
print(geneList_df3)
```

```python
################################################################################################################################
#######################################     DataFrame2    ######################################################################
################################################################################################################################

# Documentation:
    # https://readthedocs.org/projects/pyensembl/downloads/pdf/stable/
    # https://github.com/openvax/pyensembl
    # https://www.ensembl.org/info/data/ftp/index.html
    # http://ftp.ensembl.org/pub/
    
# !pip install pyensembl
# !pyensembl install --release 99 --species human

from pyensembl import EnsemblRelease

# release 99 uses human reference genome GRCh38, previously installed
data = EnsemblRelease(99)

# will return the Gene symbols
GN_df2 = ''
for e in geneList_df2:
    try:
        gene_names = data.gene_name_of_transcript_id(e)
        GN_df2 = GN_df2 + gene_names + '::' + e + ','
    except:
        GN_df2 = GN_df2 + 'Not found' + '::' + e + ','
print(GN_df2)
```

```python
################################################################################################################################
#######################################     DataFrame3    ######################################################################
################################################################################################################################

# Documentation:
    # https://readthedocs.org/projects/pyensembl/downloads/pdf/stable/
    # https://github.com/openvax/pyensembl
    # https://www.ensembl.org/info/data/ftp/index.html
    # http://ftp.ensembl.org/pub/
    
# !pip install pyensembl
# !pyensembl install --release 99 --species human

from pyensembl import EnsemblRelease

# release 99 uses human reference genome GRCh38, previously installed
data = EnsemblRelease(99)

# will return the Gene symbols
GN_df3 = ''
for e in geneList_df3:
    try:
        gene_names = data.gene_name_of_transcript_id(e)
        GN_df3 = GN_df3 + gene_names + '::' + e + ','
    except:
        GN_df3 = GN_df3 + 'Not found' + '::' + e + ','
print(GN_df3)
```

```python
################################################################################################################################
#######################################     DataFrame2    ######################################################################
################################################################################################################################

# Convert the String of gene names into a list .
# Convert this list into 2 lists and create a dictionary
# KEY = ENTS code
# VALUES = Gene NAME
geneList_df2 = GN_df2.split(",")
geneList_df2 = geneList_df2[0:(len(geneList_df2)-1)]
geneList_df2

values = []
keys = []
for e in geneList_df2:
    aux=e.split('::')
    values.append(aux[0])
    keys.append(aux[1])
#print(values)

zip_iterator = zip(keys, values) # Zip 2 list in order to generate authomatically a dictinary
dictionary_df2 = dict(zip_iterator)
dictionary_df2
```

```python
################################################################################################################################
#######################################     DataFrame3    ######################################################################
################################################################################################################################

# Convert the String of gene names into a list .
# Convert this list into 2 lists and create a dictionary
# KEY = ENTS code
# VALUES = Gene NAME
geneList_df3 = GN_df3.split(",")
geneList_df3 = geneList_df3[0:(len(geneList_df3)-1)]
geneList_df3

values = []
keys = []
for e in geneList_df3:
    aux=e.split('::')
    values.append(aux[0])
    keys.append(aux[1])
#print(values)

zip_iterator = zip(keys, values) # Zip 2 list in order to generate authomatically a dictinary
dictionary_df3 = dict(zip_iterator)
dictionary_df3
```

```python
################################################################################################################################
#######################################     DataFrame2    ######################################################################
################################################################################################################################

# Delete everything before the . in the ENST codes.
# Example: 
            # ENST00000002829.3 -->ENST00000002829
df2['Transcript'] = df2['Transcript'].str.split('.').str[0]
df2
```

```python
################################################################################################################################
#######################################     DataFrame3    ######################################################################
################################################################################################################################

# Delete everything before the . in the ENST codes.
# Example: 
            # ENST00000002829.3 -->ENST00000002829
df3['Transcript'] = df3['Transcript'].str.split('.').str[0]
df3
```

```python
################################################################################################################################
#######################################     DataFrame2    ######################################################################
################################################################################################################################

# Duplicate the column Transcript and name it Gene(s) because this column will be used with the
# list created with Gene_name:: transcript to do a replace with the Gene name.
df2['Gene(s)'] = df2['Transcript']
df2
```

```python
################################################################################################################################
#######################################     DataFrame3    ######################################################################
################################################################################################################################

# Duplicate the column Transcript and name it Gene(s) because this column will be used with the
# list created with Gene_name:: transcript to do a replace with the Gene name.
df3['Gene(s)'] = df3['Transcript']
df3
```

```python
################################################################################################################################
#######################################     DataFrame2    ######################################################################
################################################################################################################################

# Perform the subtitution Transcript to Gene NAME using the dictionary
df2 = df2.replace({"Gene(s)": dictionary_df2})
df2
```

```python
################################################################################################################################
#######################################     DataFrame3    ######################################################################
################################################################################################################################

# Perform the subtitution Transcript to Gene NAME using the dictionary
df3 = df3.replace({"Gene(s)": dictionary_df3})
df3
```

```python
################################################################################################################################
#######################################     DataFrame2 + DataFrame3    #########################################################
################################################################################################################################

df_final = df2.append(df3, ignore_index=True)
df_final = df_final.reset_index()
df_final.drop('index', axis=1, inplace=True)
df_final
```

```python
# Little comprovations of where to find 'Not found' to then drop it and reorder the index
i=df_final[df_final['Gene(s)']=='Not found']
#print(i)

df_final.drop(df_final.loc[df_final['Gene(s)']=='Not found'].index, inplace=True) # drop

df_final.index = range(len(df_final)) # reorder index
df_final
```

```python
df_final = df_final[['Gene(s)','HGVS Consequence']]
df_final
```
