```python
# Get unique gen names of clinvar file (pathogenic) to compare them with gnomad file to get genomad pathogenic
```

```python
import pandas as pd
```

```python
# Open clinvar file.
# When opening this file there's an error and uses the 1st column as the table index, so i went a little bit further to correct
# this issue. 
df_clinvar = pd.read_csv('Match(ClinVar)_(2).txt', header = 0, sep = " ")
df_clinvar = df_clinvar.reset_index()
df_clinvar.columns = [''] * len(df_clinvar.columns)
df_clinvar.columns =['Gene(s)', 'position', 'initial_aa', 'final_aa', 'na']
df_clinvar = df_clinvar[['Gene(s)', 'position', 'initial_aa', 'final_aa']]
df_clinvar = df_clinvar.drop_duplicates()
df_clinvar
```

```python
df_clinvar = df_clinvar[df_clinvar.final_aa != 'TER']
df_clinvar = df_clinvar[df_clinvar.final_aa != 'fs']
df_clinvar = df_clinvar[['Gene(s)']]
df_clinvar = df_clinvar.drop_duplicates()
df_clinvar.to_csv(r'Prot_path_clinvar.txt', header=True, index=None, sep=' ', mode='w')
df_clinvar
```

```python
# Open gnomad file and filter by these pathogenic genes
df_gnomad = pd.read_csv('Match(GnomAD)_nonPatho.txt', sep = " ")
df_gnomad = df_gnomad.reset_index()
df_gnomad.columns = ['Gene(s)', 'HGVSConsequence', 'na']
df_gnomad = df_gnomad[['Gene(s)', 'HGVSConsequence']]
df_gnomad
```

```python
result = pd.merge(df_clinvar, df_gnomad, on="Gene(s)")
result
```

```python
result.to_csv(r'Match(Gnomad)_nonPatho_Pathogenic.txt', header=True, index=None, sep=' ', mode='w')
```
