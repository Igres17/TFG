```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

```python
# Open uniprot file
df_uniprot = pd.read_csv('myOutFile_table(Uniprot).txt', header = 0, sep = " ")
df_uniprot = df_uniprot[["Gene_name", "Transmembrane_Regions", "Pfam_code"]]
df_uniprot 
```

```python
# Open clinvar file.
# When opening this file there's an error and uses the 1st column as the table index, so i went a little bit further to correct
# this issue. 
df_gnomad = pd.read_csv('Match(Gnomad)_nonPatho_Missense.txt', header = 0, sep = " ")
# df_gnomad = df_gnomad.reset_index()
# df_gnomad.columns = [''] * len(df_gnomad.columns)
# df_gnomad.columns =['Gene(s)', 'position', 'initial_aa', 'final_aa', 'na']
# df_gnomad = df_gnomad[['Gene(s)', 'position', 'initial_aa', 'final_aa']]
df_gnomad
```

```python
df_uniprot["Transmembrane_Regions"] = df_uniprot["Transmembrane_Regions"].str.split("_")
df_uniprot['Transmembrane_Regions'] = df_uniprot['Transmembrane_Regions'].fillna(0)
df_uniprot
```

```python
def list_of_lists(row, sep):
    new_l = []
    l = row["Transmembrane_Regions"]
    if l == 0:
        return [[]]
    print(l)
    for elements in l:
        s_ele = elements.split(sep)
        if s_ele not in new_l:
            new_l.append(s_ele)
    return(new_l)
    
```

```python
# llista = (df_uniprot["Transmembrane_Regions"])
# print(llista.values)
df_uniprot["Transmembrane_Regions_new"] = df_uniprot.apply(lambda row:list_of_lists(row, ".."), axis=1)
# for e in df_uniprot["Transmembrane_Regions"]:
#     df_uniprot["Transmembrane_Regions"] = pd.Series(list_of_lists(int(e), sep = ".."))
df_uniprot
```

```python
df_uniprot = df_uniprot[["Gene_name", "Transmembrane_Regions_new", "Pfam_code"]]
df_uniprot = df_uniprot.rename(columns={"Gene_name":"Gene(s)"})
df_uniprot
```

```python
result = df_gnomad.merge(df_uniprot, on='Gene(s)', how='outer')
result
```

```python
result = result.dropna()
result
```

```python
def check_transmembrane(row):
    position = int(row['position']) # Comparar si el valor es troba entre intervals,
    l_max_old = 0 # No hi ha valor old al principi
    #print(row["Transmembrane_Regions_new"])
    for i in row["Transmembrane_Regions_new"]:
        #print(i)
        try:
            l_min = i[0] 
            l_max = i[1]
        except IndexError:
            pass
        #print (position, l_min, l_max, l_max_old)
        if position >= int(l_min) and position <= int(l_max): # Es troba dintre d'un interval, no cal mirar res més.
            return "True"
        if position < int(l_min) and position > int(l_max_old): # Es troba entre intervals, és a dir, fora de TM
            return  "False" 
        l_max_old = l_max # L'últim valor del interval (l_max) pasa a ser l_max_old per veure si positions cau entre l_max_old i el següent l_min 
    return "False" # En cas que no trobi dintre de cap interval, al pasar per tots els intervals, vol dir que no és dintre d'un interval i no serà TM.

result['transmembrane'] = result.apply(lambda row:check_transmembrane(row), axis=1)
result.to_csv("TM_positions_nonPatho_missense.csv")
result[["position", "Transmembrane_Regions_new", "transmembrane"]]
```
