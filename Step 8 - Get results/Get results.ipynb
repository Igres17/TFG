```python
import pandas as pd
```

```python
# Open data into a DataFrame using Pandas
df_clinvar = pd.read_csv("pos_align_without0(Clinvar)_(2).txt", sep="\t")
df_clinvar['Database'] = 'ClinVar'
df_clinvar
```

```python
# Open data into a DataFrame using Pandas
df_gnomad = pd.read_csv("pos_align_without0(Gnomad)_nonPatho_missense.txt", sep="\t")
df_gnomad['Database'] = 'GnomAD'
df_gnomad
```

```python
# Dictionary to convert 3 letters aa to 1 letter
d = {'Cys': 'C', 'Asp': 'D', 'Ser': 'S', 'Gln': 'Q', 'Lys': 'K',
     'Ile': 'I', 'Pro': 'P', 'Thr': 'T', 'Phe': 'F', 'Asn': 'N', 
     'Gly': 'G', 'His': 'H', 'Leu': 'L', 'Arg': 'R', 'Trp': 'W', 
     'Ala': 'A', 'Val':'V', 'Glu': 'E', 'Tyr': 'Y', 'Met': 'M'}
```

```python
df_gnomad = df_gnomad.replace({"initial_aa": d})
df_gnomad = df_gnomad.replace({"final_aa": d})
df_gnomad
```

```python
# Clinvar IN Gnomad
df_final_clinvar_in_gnomad = pd.DataFrame(columns=['Gene(s)', 'position', 'initial_aa', 'final_aa', 'Transmembrane_Regions_new', 'Pfam_code', 'ID', 'ID_correct', 'pos_align', 'Database'])
for index, row in df_clinvar.iterrows():
    #print(index)
    pos = row['pos_align']
    aa_ref = row['initial_aa']
    aa_mut = row['final_aa']
    pfam = row['Pfam_code']
    #print(type(pos), type(aa_ref), type(aa_mut))
    sol = df_gnomad[(df_gnomad["initial_aa"] == aa_ref) & (df_gnomad["final_aa"] == aa_mut) & (df_gnomad["pos_align"] == pos)  & (df_gnomad["Pfam_code"] == pfam)]
    if not sol.empty:
        #print(sol)
        df_final_clinvar_in_gnomad = df_final_clinvar_in_gnomad.append(row)
        df_final_clinvar_in_gnomad = df_final_clinvar_in_gnomad.append(sol)
df_final_clinvar_in_gnomad.reset_index(drop = True, inplace = True)
df_final_clinvar_in_gnomad
```

```python
# Clinvar IN Clinvar
df_final_clinvar_in_clinvar = pd.DataFrame(columns=['Gene(s)', 'position', 'initial_aa', 'final_aa', 'Transmembrane_Regions_new', 'Pfam_code', 'ID', 'ID_correct', 'pos_align', 'Database'])
df_clinvar = df_clinvar.drop_duplicates()
for index, rows in df_clinvar.iterrows():
    #print(index)
    position = rows['pos_align']
    aa_i = rows['initial_aa']
    aa_f = rows['final_aa']
    pfam_code = rows['Pfam_code']
    #print(type(pos), type(aa_ref), type(aa_mut))
    for e in position:
        for i in pfam_code:
            solution = df_clinvar[(df_clinvar["initial_aa"] == aa_i) & (df_clinvar["final_aa"] == aa_f) & (df_clinvar["pos_align"] == position)  & (df_clinvar["Pfam_code"] == pfam_code)]
            if len(solution)>1:
        #print(solution)
        #df_final_clinvar_in_clinvar = df_final_clinvar_in_clinvar.append(rows)
                df_final_clinvar_in_clinvar = df_final_clinvar_in_clinvar.append(solution)
df_final_clinvar_in_clinvar.reset_index(drop = True, inplace = True)
df_final_clinvar_in_clinvar = df_final_clinvar_in_clinvar.drop_duplicates()
df_final_clinvar_in_clinvar
```

```python
# Gnomad IN Gnomad
from datetime import datetime
init_time = datetime.now()

df_final_gnomad_in_gnomad = pd.DataFrame(columns=['Gene(s)', 'position', 'initial_aa', 'final_aa', 'Transmembrane_Regions_new', 'Pfam_code', 'ID', 'ID_correct', 'pos_align', 'Database'])
df_gnomad = df_gnomad.drop_duplicates()
for index, files in df_gnomad.iterrows():
    print((int(index)/len(df_gnomad))*100)
    #print(index)
    position_g = files['pos_align']
    aa_i_g = files['initial_aa']
    aa_f_g = files['final_aa']
    pfam_code_g = files['Pfam_code']
    #print(type(pos), type(aa_ref), type(aa_mut))
    for e in position_g:
        for i in pfam_code_g:
            solution = df_gnomad[(df_gnomad["initial_aa"] == aa_i_g) & (df_gnomad["final_aa"] == aa_f_g) & (df_gnomad["pos_align"] == position_g)  & (df_gnomad["Pfam_code"] == pfam_code_g)]
            if len(solution)>1:
        #print(solution)
        #df_final_clinvar_in_clinvar = df_final_clinvar_in_clinvar.append(rows)
                df_final_gnomad_in_gnomad = df_final_gnomad_in_gnomad.append(solution)
df_final_gnomad_in_gnomad.reset_index(drop = True, inplace = True)
df_final_gnomad_in_gnomad = df_final_gnomad_in_gnomad.drop_duplicates()
df_final_gnomad_in_gnomad

fin_time = datetime.now()
print("Execution time (for loop): ", (fin_time-init_time))
```
