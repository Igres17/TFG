```python
import pandas as pd
```

```python
df = pd.read_csv("df_final_clinvar_in_gnomad_nonPatho.txt", sep="\t", header=0)
df= df.drop_duplicates()
df
```

```python
df[df['Gene(s)']== 'SCN10A']
```

```python
df_filter = pd.DataFrame(columns=df.columns)
df_nofilter = pd.DataFrame(columns=df.columns)
for index, row in df.iterrows():
    gens = row['Gene(s)']
    pos = row['position']
    aa_i = row['initial_aa']
    aa_f = row['final_aa']
    #database = row['Database']
    df_search = df[(df['Gene(s)']==gens) & (df['position']==pos) & (df['initial_aa']==aa_i) & (df['final_aa']==aa_f)]
    if len(df_search['Database'].unique())<2:
        #print(df_search)
        df_filter = pd.concat([df_filter, df_search])
    else:
        df_nofilter = pd.concat([df_nofilter, df_search])
df_filter   
```

```python
blacklist = df_nofilter['Gene(s)'].unique().tolist()
len(blacklist)
```

```python
df_filter = df_filter[df_filter['Gene(s)'].isin(blacklist)]
df_filter = df_filter.reset_index()
df_filter.drop ('index', axis=1, inplace=True) 
df_filter
```

```python
df_filter.to_csv('df_clinvar_in_gnomad_without_homologous_nonPatho.txt', header=True, index=False, sep='\t', mode='w')
```
