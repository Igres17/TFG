```python
import pandas as pd
```

```python
####CLINVAR IN CLINVAR
```

```python
df_clinvar_in_clinvar = pd.read_csv("df_final_clinvar_in_clinvar_(2).txt", sep="\t", header=0)
df_clinvar_in_clinvar = df_clinvar_in_clinvar.drop_duplicates()
df_clinvar_in_clinvar
```

```python
dups_color_clinvar_in_clinvar = df_clinvar_in_clinvar.pivot_table(columns=['initial_aa', 'final_aa', 'Pfam_code', 'pos_align'], aggfunc='size')
dups_color_clinvar_in_clinvar.to_csv('PAIRS_CLINVAR IN CLINVAR', header=True, index=False, sep='\t', mode='w')
dups_color_clinvar_in_clinvar
```

```python
df_clinvar_in_clinvar_SUM = pd.read_csv("PAIRS_CLINVAR IN CLINVAR", sep="\t", header=0)
df_clinvar_in_clinvar_SUM = df_clinvar_in_clinvar_SUM.rename(columns={"0":"Values"})
df_clinvar_in_clinvar_SUM
```

```python
df_clinvar_in_clinvar_SUM.pivot_table(columns=['Values'], aggfunc='size')
```
