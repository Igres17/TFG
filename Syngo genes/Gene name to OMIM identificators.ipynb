```python
# This programs allows to convert the gene name into the OMIM identificator. 
# This is used to create links to then do text mining into the OMIM database
```

```python
import pandas as pd
```

```python
# Open ClinVar file
CountsFile1 = pd.read_excel('Genes_patho_length.xls')
CountsFile = CountsFile1[['Gene(s)']]
CountsFile
```

```python
# A tab-delimited file containing OMIM's Synopsis of the Human Gene Map including additional information such as genomic coordinates and inheritance.
OMIMfile = pd.read_csv("mim2gene.txt", header=0, skiprows=[0, 1, 2, 3], sep='\t')
OMIMfile = OMIMfile[['# MIM Number', 'Approved Gene Symbol (HGNC)']]
OMIMfile.rename(columns = {'Approved Gene Symbol (HGNC)':'Gene(s)'}, inplace = True)
OMIMfile
```

```python
OMIMfile = OMIMfile.dropna()
OMIMfile['# MIM Number'] = OMIMfile['# MIM Number'].astype('string')
OMIMfile
```

```python
OMIMfile['# MIM Number'] = OMIMfile.groupby(['Gene(s)'])['# MIM Number'].transform(lambda x : ','.join(x))
OMIMfile
```

```python
merged = pd.merge(CountsFile, OMIMfile, on='Gene(s)')
merged
```

```python
final = pd.merge(CountsFile1, merged, on='Gene(s)', how='outer')
final[0:10]
```

```python
final.to_csv('counts_to_MIMcode.txt', header=True, index=None, sep=' ', mode='w')
```
