```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

```python
# Open ClinVar file
ClinVar = pd.read_csv("File2_tojoinwith_GnomAD(ClinVar).txt", sep=" ", header=0)
ClinVar
```

```python
# Select the wanted columns ([['Gene(s)', 'Molecular Consequence', 'Database']]) and
# rename the 'Molecular Consequence' column into 'Molecular Consequence/VEP Annotation' in order
# to have the same names with the GnomAD file to join them.
df_Clinvar = ClinVar[['Gene(s)', 'Molecular Consequence', 'Database']]
df_Clinvar.rename(columns={'Molecular Consequence': 'Molecular Consequence/VEP Annotation'}, inplace=True)
df_Clinvar
```

```python
# Create a table of counts as a df ( the gene names will desappear when exported, need to be added
# in the aftermath)
# A total column of the total number of molecular consequences in the clinVar file is added
df2 = df_Clinvar.reset_index().groupby(['Gene(s)', 'Molecular Consequence/VEP Annotation']).count().unstack('Molecular Consequence/VEP Annotation').fillna(0)
df2.columns = df2.columns.droplevel(0)

df2.columns = [x for x in df2.columns]
df2 = df2.iloc[: , :4]

df2.rename(columns={'Frameshift': 'Frameshift_ClinVar', 'Missense':'Missense_ClinVar', 'Nonsense': 'Nonsense_ClinVar', 'Splicesites': 'Splicesites_ClinVar'}, inplace=True)

df2['Gene(s)'] = df2.index

df2 = df2[['Gene(s)', 'Frameshift_ClinVar', 'Missense_ClinVar', 'Nonsense_ClinVar', 'Splicesites_ClinVar']]

df2.reset_index(drop=True, inplace=True)

df2
```

```python
# Open de GnomAD file as df
GnomAD = pd.read_csv("File2_tojoinwith_ClinVar(GnomAD).txt", sep=" ", header=0)
GnomAD
```

```python
# Select the wanted columns ([['Gene(s)', 'VEP Annotation', 'Database']]) and
# rename the ''VEP Annotation'' column into 'Molecular Consequence/VEP Annotation' in order
# to have the same names with the GnomAD file to join them.
df_Gnomad = GnomAD[['Gene(s)', 'VEP Annotation', 'Database']]
df_Gnomad.rename(columns={'VEP Annotation': 'Molecular Consequence/VEP Annotation'}, inplace=True)
df_Gnomad
```

```python
# A little comprovation of how much Molecular consequences are in the GnomAD file ordered by name
a = df_Gnomad['Molecular Consequence/VEP Annotation'].unique()
print(sorted(a))
```

```python
# Create a table of counts as a df ( the gene names will desappear when exported, need to be added
# in the aftermath)
# A total column of the total number of molecular consequences in the GnomAD file is added
df3 = df_Gnomad.reset_index().groupby(['Gene(s)', 'Molecular Consequence/VEP Annotation']).count().unstack('Molecular Consequence/VEP Annotation').fillna(0)
df3.columns = df3.columns.droplevel(0)

df3.columns = [x for x in df3.columns]
df3 = df3.iloc[: , :19]

df3.rename(columns={'3_prime_UTR_variant': '3_prime_UTR_GnomAD', '5_prime_UTR_variant':'5_prime_UTR_GnomAD', 'coding_sequence_variant': 'coding_sequence_GnomAD', 'frameshift_variant': 'frameshift_GnomAD', 'incomplete_terminal_codon_variant':'incomplete_terminal_codon_GnomAD', 'inframe_deletion':'inframe_deletion_GnomAD', 'inframe_insertion':'inframe_insertion_GnomAD', 'intron_variant':'intron_GnomAD', 'missense_variant':'missense_GnomAD', 'non_coding_transcript_exon_variant':'non_coding_transcript_exon_GnomAD', 'protein_altering_variant':'protein_altering_GnomAD', 'splice_acceptor_variant':'splice_acceptor_GnomAD', 'splice_donor_variant':'splice_donor_GnomAD', 'splice_region_variant':'splice_region_GnomAD', 'start_lost':'start_lost_GnomAD', 'stop_gained':'stop_gained_GnomAD', 'stop_lost':'stop_lost_GnomAD', 'stop_retained_variant':'stop_retained_GnomAD', 'synonymous_variant':'synonymous_GnomAD'}, inplace=True)

df3['Gene(s)'] = df3.index

df3 = df3[['Gene(s)', '3_prime_UTR_GnomAD', '5_prime_UTR_GnomAD', 'coding_sequence_GnomAD', 'frameshift_GnomAD', 'incomplete_terminal_codon_GnomAD', 'inframe_deletion_GnomAD', 'inframe_insertion_GnomAD', 'intron_GnomAD', 'missense_GnomAD', 'non_coding_transcript_exon_GnomAD', 'protein_altering_GnomAD', 'splice_acceptor_GnomAD', 'splice_donor_GnomAD', 'splice_region_GnomAD', 'start_lost_GnomAD', 'stop_gained_GnomAD', 'stop_lost_GnomAD', 'stop_retained_GnomAD', 'synonymous_GnomAD']]

df3.reset_index(drop=True, inplace=True)

df3
```

```python
# Then both ClinVar and GnomAD dataframes are concatenated into another dataframe. Another count
# is added in order to sum the TOTAL molecular consequences in both the clinVar and the Gnomad files.
mergedRes = pd.concat([df2, df3], ignore_index=True)
mergedRes = mergedRes.replace(np.nan, 0)

mergedRes
```

```python
columns = list(mergedRes.columns.values)
columns = columns[1::]
columns
```

```python
#(mergedRes.groupby(['Gene(s)']).agg({'Frameshift_ClinVar': 'sum','Missense_ClinVar': 'sum','Nonsense_ClinVar': 'sum', 'totals_ClinVar': 'sum', '3_prime_UTR_GnomAD': 'sum', '5_prime_UTR_GnomAD': 'sum', 'coding_sequence_GnomAD': 'sum', 'frameshift_GnomAD': 'sum', 'incomplete_terminal_codon_GnomAD': 'sum', 'inframe_deletion_GnomAD': 'sum', 'inframe_insertion_GnomAD': 'sum', 'intron_GnomAD': 'sum', 'missense_GnomAD': 'sum', 'non_coding_transcript_exon_GnomAD': 'sum', 'protein_altering_GnomAD': 'sum', 'splice_acceptor_GnomAD': 'sum', 'splice_donor_GnomAD': 'sum', 'splice_region_GnomAD': 'sum', 'start_lost_GnomAD': 'sum', 'stop_gained_GnomAD': 'sum', 'stop_lost_GnomAD': 'sum', 'stop_retained_GnomAD': 'sum', 'synonymous_GnomAD': 'sum', 'totals_GnomAD': 'sum'})
sumRes = mergedRes.groupby(['Gene(s)'])['Frameshift_ClinVar',
 'Missense_ClinVar',
 'Nonsense_ClinVar',
 'Splicesites_ClinVar',
 '3_prime_UTR_GnomAD',
 '5_prime_UTR_GnomAD',
 'coding_sequence_GnomAD',
 'frameshift_GnomAD',
 'incomplete_terminal_codon_GnomAD',
 'inframe_deletion_GnomAD',
 'inframe_insertion_GnomAD',
 'intron_GnomAD',
 'missense_GnomAD',
 'non_coding_transcript_exon_GnomAD',
 'protein_altering_GnomAD',
 'splice_acceptor_GnomAD',
 'splice_donor_GnomAD',
 'splice_region_GnomAD',
 'start_lost_GnomAD',
 'stop_gained_GnomAD',
 'stop_lost_GnomAD',
 'stop_retained_GnomAD',
 'synonymous_GnomAD'].apply(lambda x : x.astype(int).sum())

sumRes['Gene(s)'] = sumRes.index

sumRes = sumRes[['Gene(s)', 'Frameshift_ClinVar', 'Missense_ClinVar', 'Nonsense_ClinVar', 'Splicesites_ClinVar', '3_prime_UTR_GnomAD', '5_prime_UTR_GnomAD', 'coding_sequence_GnomAD', 'frameshift_GnomAD', 'incomplete_terminal_codon_GnomAD', 'inframe_deletion_GnomAD', 'inframe_insertion_GnomAD', 'intron_GnomAD', 'missense_GnomAD', 'non_coding_transcript_exon_GnomAD', 'protein_altering_GnomAD', 'splice_acceptor_GnomAD', 'splice_donor_GnomAD', 'splice_region_GnomAD', 'start_lost_GnomAD', 'stop_gained_GnomAD', 'stop_lost_GnomAD', 'stop_retained_GnomAD', 'synonymous_GnomAD']]

sumRes.reset_index(drop=True, inplace=True)

sumRes
```

```python
# Per fer el count
remove_Gnomad = ['Gene(s)','3_prime_UTR_GnomAD', '5_prime_UTR_GnomAD', 'coding_sequence_GnomAD', 'frameshift_GnomAD', 'incomplete_terminal_codon_GnomAD', 'inframe_deletion_GnomAD', 'inframe_insertion_GnomAD', 'intron_GnomAD', 'missense_GnomAD', 'non_coding_transcript_exon_GnomAD', 'protein_altering_GnomAD', 'splice_acceptor_GnomAD', 'splice_donor_GnomAD', 'splice_region_GnomAD', 'start_lost_GnomAD', 'stop_gained_GnomAD', 'stop_lost_GnomAD', 'stop_retained_GnomAD', 'synonymous_GnomAD']
dfaux_CLINVAR = sumRes[sumRes.columns.difference(remove_Gnomad)]
#dfaux_CLINVAR


remove_Clinvar = ['Gene(s)','Frameshift_ClinVar', 'Missense_ClinVar', 'Nonsense_ClinVar', 'Splicesites_ClinVar']
dfaux_GNOMAD = sumRes[sumRes.columns.difference(remove_Clinvar)]
dfaux_GNOMAD
# dfaux_GNOMAD = dfaux.drop(, axis=1)

totals = dfaux_CLINVAR.sum(axis=1)
totals.name = 'totals_ClinVar'

totals2 = dfaux_GNOMAD.sum(axis=1)
totals2.name = 'totals_GnomAD'

totals3 = sumRes.sum(axis=1)
totals3.name = 'TOTALS'

sumRes = pd.concat([sumRes, totals, totals2, totals3], axis=1)

sumRes = sumRes[['Gene(s)', 'Frameshift_ClinVar', 'Missense_ClinVar', 'Nonsense_ClinVar', 'Splicesites_ClinVar', 'totals_ClinVar', '3_prime_UTR_GnomAD', '5_prime_UTR_GnomAD', 'coding_sequence_GnomAD', 'frameshift_GnomAD', 'incomplete_terminal_codon_GnomAD', 'inframe_deletion_GnomAD', 'inframe_insertion_GnomAD', 'intron_GnomAD', 'missense_GnomAD', 'non_coding_transcript_exon_GnomAD', 'protein_altering_GnomAD', 'splice_acceptor_GnomAD', 'splice_donor_GnomAD', 'splice_region_GnomAD', 'start_lost_GnomAD', 'stop_gained_GnomAD', 'stop_lost_GnomAD', 'stop_retained_GnomAD', 'synonymous_GnomAD', 'totals_GnomAD', 'TOTALS']]

sumRes
```

```python
# This table is exported into a CSV. Remember that with this .csv the gene name will disappear.
# Next, the gene names will be added
sumRes.to_csv('File1_v3(Counts).txt', header=True, index=None, sep=' ', mode='a')
```
