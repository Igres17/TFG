```python
import pandas as pd
pd.get_option("display.max_columns", None)
```

```python
# Open the GnomAD combined files and delete which it is duplicated (the header was
# duplicated because we joined all files using git-bash)

df = pd.read_csv('combined.csv')
df = df.drop_duplicates()
i=df[df['Chromosome']=='Chromosome']
#print(i)
df = df.drop(df.index[1279])
df
```

```python
# Create a df with the desired columns
df2 = df[['Transcript','HGVS Consequence','VEP Annotation']]
df2[0:40]
```

```python
# Create a string with all the unique ENST codes without the extension (.number) in order to convert
# them into the Gene Name using a function.
u_transcripts = df2['Transcript'].unique().tolist()
#u_transcripts
sep = '.'
stripped =''
for e in u_transcripts:
    stripped = stripped + ',' +(e.split(sep, 1)[0])
stripped = stripped [1::]
print(stripped)
```

```python
geneList = list(stripped.split(',')) 
print(geneList)
```

```python
# Documentation:
    # https://readthedocs.org/projects/pyensembl/downloads/pdf/stable/
    # https://github.com/openvax/pyensembl
    # https://www.ensembl.org/info/data/ftp/index.html
    # http://ftp.ensembl.org/pub/
    
# !pip install pyensembl
# !pyensembl install --release 99 --species human

from pyensembl import EnsemblRelease

# release 99 uses human reference genome GRCh38, previously installed
data = EnsemblRelease(99)

# will return the Gene symbols
GN = ''
for e in geneList:
    try:
        gene_names = data.gene_name_of_transcript_id(e)
        GN = GN + gene_names + '::' + e + ','
    except:
        GN = GN + 'Not found' + '::' + e + ','
print(GN)
```

```python
# Convert the String of gene names into a list .
# Convert this list into 2 lists and create a dictionary
# KEY = ENTS code
# VALUES = Gene NAME
geneList = GN.split(",")
geneList = geneList[0:(len(geneList)-1)]
geneList

values = []
keys = []
for e in geneList:
    aux=e.split('::')
    values.append(aux[0])
    keys.append(aux[1])
#print(values)

zip_iterator = zip(keys, values) # Zip 2 list in order to generate authomatically a dictinary
dictionary = dict(zip_iterator)
dictionary
```

```python
# Delete everything before the . in the ENST codes.
# Example: 
            # ENST00000002829.3 -->ENST00000002829
df2['Transcript'] = df2['Transcript'].str.split('.').str[0]
df2
```

```python
# Duplicate the column Transcript and name it Gene(s) because this column will be used with the
# list created with Gene_name:: transcript to do a replace with the Gene name.
df2['Gene(s)'] = df2['Transcript']
df2
```

```python
# Reorder the columns to have the same format as the obtained with the ClinVar file
df2 = df2[["Transcript", "Gene(s)", "HGVS Consequence", "VEP Annotation"]]
df2
```

```python
# Perform the subtitution Transcript to Gene NAME using the dictionary
df2 = df2.replace({"Gene(s)": dictionary})
df2
```

```python
# Little comprovations of where to find 'Not found' to then drop it and reorder the index
i=df2[df2['Gene(s)']=='Not found']
#print(i)

df2.drop(df2.loc[df2['Gene(s)']=='Not found'].index, inplace=True) # drop

df2.index = range(len(df2)) # reorder index

df2[1850:1900]
```

```python
# Add The Database and the Clinical significance
df2["Clinical Significance"] = 'Non pathogenic'
df2["Database"] = 'GnomAD'
df2
```

```python
df2.to_csv('File2(GnomAD).txt', header=True, index=None, sep=' ', mode='a')
```

```python
# At the end we're going to use this columns in order to join the file with the ClinVar ones in order
# to get the file of counts.
df2_2 = df2[["Gene(s)", "HGVS Consequence", "VEP Annotation", "Clinical Significance", "Database"]]
df2_2.to_csv('File2_tojoinwith_ClinVar(GnomAD).txt', header=True, index=None, sep=' ', mode='a')
```
